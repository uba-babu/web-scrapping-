```python
url = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='
prices_final = []
for i in range(2,11):
    link = url + str(i)
    print(f'scrapping currently page number -->{i}\n\n\n\n')
    conn = requests.get(link)
    content = conn.content 
    soup = BeautifulSoup(content,'html.parser')
    prices = soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
    
    for i in prices:
        prices_final.append(i.text)
```
